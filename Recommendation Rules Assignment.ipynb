```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
book=pd.read_csv("C:/Users/Vijay Rathod/Downloads/Book.csv",encoding='latin1')
book
```

```python
### Removing unuseble col
del book['Unnamed: 0']
```

```python
#### Renameing the columns
book1=book.rename({'User.ID':'userid','Book.Title':'booktitle','Book.Rating':'bookrating'},axis=1)
book1
```

```python
### Seeing number of Unique Values in data set
len(book1.userid.unique())
```

```python
len(book1.booktitle.unique())
```

```python
#### EDA
```

```python
book1.info()
```

```python
book_df=book1.pivot_table(index='userid',columns='booktitle',values='bookrating').reset_index(drop= True)
book_df
```

```python
# Let's Set unique user-id's from original dataset as index for new created pivot matrix
book_df.index = book1.userid.unique()
book_df
```

```python
### Imputing NaN values with 0
book_df.fillna(0,inplace= True)
```

```python
book_df.head()
```

```python
### Calculating Cosine Similarity between
user_sim=1-pairwise_distances(book_df.values,metric='cosine')
user_sim
```

```python
### Store the data in a data frame
userdf=pd.DataFrame(user_sim)
userdf
```

```python
### Set the index and coiumn and name to user ids
userdf.index=book1.userid.unique()
userdf.columns=book1.userid.unique()
```

```python
userdf.iloc[0:5,0:5]
```

```python
# Since Correlation with self is always 1, replacing diagonal values of correlation matrix with zeros
np.fill_diagonal(user_sim, 0)
userdf.iloc[0:5, 0:5]
```

```python
#### Most similar Users
userdf.idxmax(axis=1)[0:30]

#### this shows us highly similar user
```
